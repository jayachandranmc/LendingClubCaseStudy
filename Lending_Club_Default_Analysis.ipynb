```python
#Data Cleaning
print("Data cleaning")
# loan.columns
unnecessary_columns = ["id","member_id"]
loan.drop(unnecessary_columns,axis=1,inplace=True)
loan.columns
```

```python
# Data cleaning
# round((loan.isnull().sum()/loan.shape[0])*100,2)
# emp_length: Employment length in years. Possible values are between 0 and 10 
#   where 0 means less than one year and 10 means ten or more years. 
print("remove null and NA values")
loan.dropna(subset = ['emp_length'], inplace=True)
round((loan.isnull().sum()/loan.shape[0])*100,2)
```

```python
# Data Cleaning
# Interest rate column has % symbol, need to remove it, so that we can make a graph out of the numerical value
loan["int_rate"] = loan['int_rate'].apply(lambda x: float(x.split("%")[0]))
loan["int_rate"]
```

```python
loan['emp_length']
```

```python
# There is a need to remove non numeric characters from the emp_length, so that we can use this column for analysis
# For example years, < symbol, and also n/a. 
# Verify the csv file by saving as xls and do a pivot table to keep validating the data parallely and confirming. 

loan['emp_length'] = loan['emp_length'].astype("str")
loan['emp_length'] = loan['emp_length'].apply(lambda x: x.replace("years",""))
loan['emp_length'] = loan['emp_length'].apply(lambda x: x.replace("year",""))
loan['emp_length'] = loan['emp_length'].apply(lambda x: x.replace("10+","11"))
loan['emp_length'] = loan['emp_length'].apply(lambda x: x.replace("< 1","0"))
```

```python
# Verify the employment length column
loan['emp_length'].value_counts()
# output confirms that, the employment length column is cleaned up.
```

```python
# Below is option code to covert the type of employment length to integer. Needed for analysis
loan['emp_length']=loan['emp_length'].astype('int')
loan['emp_length'].value_counts()
```

```python
# Covert the annual income to categorical values 

#low-income: 0-50000
#mid-income: 50001-100000
#high-income: 100001>
def annul_inc_conversion(x):
  if x<= 50000:
    return "low income"
  elif x> 50000 and x <= 100000:
    return "medium income"
  else:
    return "high income"
```

```python
loan["annual_inc"] = loan["annual_inc"].apply(lambda x:annul_inc_conversion(x))
```

```python
loan["annual_inc"]
```

```python
# Convert interest rate column to categorical values
#low-intersrt: 7.5
#mid-interst: 7.6-13
#high-insterst: 13+
def interst_rate_conversion(x):
  if x<= 7.5:
    return "low interst rate"
  elif x>7.5  and x <= 13.0:
    return "medium interst rate"
  else:
    return "high interst rate"
```

```python
loan["int_rate"]= loan["int_rate"].apply(lambda x:interst_rate_conversion(x) )
```
